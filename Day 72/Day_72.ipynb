```python
substract_column=clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, "Difference", substract_column)
```

```python
clean_df.tail()
```

```python
low_risk = clean_df.sort_values('Difference')
low_risk[['Undergraduate Major', 'Difference']].head()
```

Using the .sort_values() method, can you find the degrees with the highest potential? Find the top 5 degrees with the highest values in the 90th percentile.

```python
highest_salaries = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False).head()
print(highest_salaries[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']])
```

Also, find the degrees with the greatest spread in salaries. Which majors have the largest difference between high and low earners after graduation.

```python
highest_difference=clean_df.sort_values("Difference", ascending=False)
print(highest_difference[['Undergraduate Major', 'Difference']].head())
```
